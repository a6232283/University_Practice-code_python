```python
#端口掃描

ports={'21','80','22','8080','3306','3389','443','445'}
sever=socket.socket(socket.AF_INET, socket.SOCK_STREAM)
for port in ports:
    result=sever.connect_ex(('www.google.com',int(port)))
    if result==0:
        print(port+'open')
    else:
        print(port+'close')
```

```python
def suburl(url):
    urls=url.replace('www','')
    for Subdomains in open('subnames_full.txt'):
        Subdomains=Subdomains.replace('\n','')
        url=Subdomains+urls

        try:
            ip=socket.gethostbyname(url)
            print(url+' -> '+ip)
            time.sleep(0.1)
        except Exception as e:
            pass
    
if __name__ == '__main__':
    inurl=input('請輸入域名')
    suburl(inurl)
```

```python
#根目錄路掃描

def inhttp(http):
    url='http://'+http
    asp_file=open('asp.txt','r')
    for asp in asp_file:
        asp=asp.replace('\n','')
        urls=url+asp
        http=requests.get(urls).status_code

        if http==200:
            print(urls+' -> '+str(http))
        else:
            pass

if __name__ == '__main__':
    http=input('請輸入網址')
    inhttp(http)
```
